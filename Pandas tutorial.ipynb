## Creating a Series
- Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.)

```python
import pandas as pd 
import numpy as np 


# Creating empty series 
ser = pd.Series() 

print(ser) 

# simple array 
data = np.array(['c', 'o', 'r', 'o', 'n', 'a']) 

ser = pd.Series(data) 
print(ser) 
```

```python
ser[2] # retreiving a data as per index
```

## Selecting and Retrieving

```python
ser = pd.Series(['c', 'o', 'r', 'o', 'n', 'a'], index = ['row 1','row 2','row 3','row 4','row 5','row 6'])
ser
```

```python
# Selecting values by label
ser['row 2']
```

```python
ser_1 = pd.Series([1,2,3])
print(ser_1)
print('\n')
# row label assignment with index
ser_1.index = ['row 1','row 2','row 3']
print(ser_1)
print('\n')
#label assignment during initiation
ser_2 = pd.Series([7,8,9,10],index = ['row 1','row 2','row 3', 'row 4'])
print(ser_2)
print('\n')
# retreiving multiple value with label as reference
ser_3 = pd.Series(np.arange(8), index = ['row 1','row 2','row 3','row 4','row 5','row 6','row 7','row 8'])
print(ser_3)
print('\n')

ser_2[['row 2','row 3','row 1']]

# retreive with index through slicing
ser_3[3:6] # 3 is inclusive and 6 is exclusive

# retreive record by label with slicing
ser_3['row 2':'row 4'] # here 'row 4' s inclusive
```

## Creating a DataFrame
- Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.
- In the real world, a Pandas DataFrame will be created by loading the datasets from existing storage, storage can be SQL Database, CSV file, and Excel file. Pandas DataFrame can be created from the lists, dictionary, and from a list of dictionary etc.

```python
import pandas as pd 

# Calling DataFrame constructor 
df = pd.DataFrame() 
print(df) 

# list of strings 
lst = ['a', 'hot', 'programmer', 'girlfriend', 'is', 'still', 'a myth'] 

# Calling DataFrame constructor on list 
df = pd.DataFrame(lst) 
print(df) 

#Assign labels to rows
df.index = ['row 1','row 2','row 3','row 4','row 5','row 6','row 7']
df

# Assigning lables to columns
df.columns = ['only column']
df
```

## Creating DataFrame from dict of ndarray/lists
- To create DataFrame from dict of narray/list, all the narray must be of same length

```python
import pandas as pd
 
# intialise data of lists.
data = {'Name':['J Hinton', 'Yann LeCun', 'Andrew Ng', 'E Musk'],
        'Age':[20, 21, 19, 18]}
 
# Create DataFrame
df = pd.DataFrame(data)
 
# Print the output.
print(df)
```

## Column Selection
- In Order to select a column in Pandas DataFrame, we can either access the columns by calling them by their columns names

```python
# Define a dictionary containing employee data
data = {'Name':['Tom', 'Jerry', 'Iron Man', 'Thor'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Mumbai', 'Calicut', 'Bengaluru'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}
 
# Convert the dictionary into DataFrame 
df = pd.DataFrame(data)
print(df)
print('\n')
# select two columns
print(df[['Name', 'Qualification']])
```

## Row Selection
- Pandas provide a unique method to retrieve rows from a Data frame. DataFrame.loc[] method is used to retrieve rows from Pandas DataFrame. Rows can also be selected by passing integer location to an iloc[] function.

- lets consider a mtcars.csv dataset for this example


Motor Trend Car Road Tests
- Description
The data was extracted from the 1974 Motor Trend US magazine, and comprises fuel consumption and 10 aspects of automobile design and performance for 32 automobiles (1973–74 models).

- A data frame with 32 observations on 11 (numeric) variables.

    - mpg	- Miles/(US) gallon
    - cyl	- Number of cylinders
    - disp	- Displacement (cu.in.)
    - hp	- Gross horsepower
    - drat	- Rear axle ratio
    - wt	- Weight (1000 lbs)
    - qsec	- 1/4 mile time
    - vs	- Engine (0 = V-shaped, 1 = straight)
    - gear	- Number of forward gears
    - carb	- Number of carburetors
 

```python
import pandas as pd
data = pd.read_csv('cars.csv', index_col = 'car_model')
data.head(10)
```

```python
# Indexing a DataFrame using .loc['']
# This function selects data by the label of the rows and columns. 
first = data.loc['Datsun 710']
second = data.loc['Valiant']

print(first, '\n', second)
```

```python
# Indexing a DataFrame using .iloc[ ]
# This function allows us to retrieve rows and columns by position. 
# In order to do that, we’ll need to specify the positions of #the rows that we want, 
# and the positions of the columns that we want as well.

row2 = data.iloc[3] # here we have the details of 'Hornet 4 Drive'
row2
```

```python
# retrieve multiple columns
data.iloc[1:5,1:7]
```

## Missing Data
Missing Data can occur when no information is provided for one or more items or for a whole unit. Missing Data is a very big problem in real life scenario. Missing Data can also refer to as NA(Not Available) values in pandas


- In order to fill null values in a datasets, we use fillna(), replace() and interpolate() function these function replace NaN values with some value of their own. All these function help in filling a null values in datasets of a DataFrame. Interpolate() function is basically used to fill NA values in the dataframe but it uses various interpolation technique to fill the missing values rather than hard-coding the value.

- lets take a new dataset for this. its the same as the previous 'cars.csv dataset', however different in terms of certain missing values and NaN values. We'll try and make the dataset as clean as possible basedon the 'domain knowledge'.

```python
# now that we findout that we need to replace all these special caracters ('$','?','*','-') with NaN.
missing_data.replace(['$','?','*','-'],np.nan,inplace = True)
missing_data
```

```python
# now we need to treat the outlier value '999' in the column 'mpg'
missing_data.loc[missing_data.mpg > 100, 'mpg'] = np.nan # i.e, if there is any value > 100, those values need to replaced by NaN
missing_data
```
